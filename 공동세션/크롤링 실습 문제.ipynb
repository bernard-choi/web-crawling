```python
#후손선택자를 이용해서 가져오려고 했으나 h4가 바로 닫힌다... 다른 방법을 찾아야한다.
# a태그의 href속성값을 이용해보자!
```

```python
lists = ['mon','tue','wed','thu','fri','sat','sun']
```

```python
yunsiks = soup.select("a.title")

for i in range(len(lists)):
    print(lists[i])
    for yunsik in yunsiks:
        if yunsik.attrs[''].split("")[] == lists[i]:
            print (yunsik.text)
```
